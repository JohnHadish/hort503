```python
import pandas as pd
```

```python
PI = pd.read_csv("PI_DataSet.txt", sep="\t")
PI.head()
```

```python
PI_AA = PI.drop(PI.columns[1:9], axis=1)
PI_AA.head()
```

```python
PI_AA_Melted = pd.melt(PI_AA,
                        id_vars=["SeqID"],
                        var_name="AA_Pos",
                        value_name="AA")
PI_AA_Melted["AA_Pos"] = PI_AA_Melted["AA_Pos"].str.extract('(\d+)', expand=False).astype("int")
PI_AA_Melted.head()
```

```python
PI_AA_Melted = PI_AA_Melted.sort_values(["SeqID", "AA_Pos"])
PI_AA_Melted = PI_AA_Melted.reset_index(drop=True)
PI_AA_Melted.head()
```
