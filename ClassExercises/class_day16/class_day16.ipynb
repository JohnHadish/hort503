```python
import pandas as pd
import numpy as np
```

```python
#First I imported the corn csv
corn = pd.read_csv("corn.csv")

#Then I created titles for each year on the table. I also included one for "State"
years = list(range(2000,2018))
print(years)
colName = ["State"]
colName.extend(years)

#Segment out the part of the table and assign column names
corn_bt = corn.iloc[3:18, 0:19]
corn_bt.columns =colName

#This part is goofy and terrible because pandas is a little bit finiky. 
#This establishes the ranges we want to use
a = range(19,37)
b = [0]
b.extend(a)
print(b)

#This is the part that takes the table for herbicides
corn_herbicide = corn.iloc[3:18, b]
corn_herbicide.columns = colName

corn_bt
```

```python
corn_herbicide
```

```python
corn_bt_tidy = pd.melt(corn_bt,
                    id_vars=['State'],
                    var_name="Years",
                    value_name="Percent of Crop").dropna()
corn_bt_tidy.head()
```

```python
corn_bt_tidy["Crop"] = ("Corn")
corn_bt_tidy["Variety"] = "Produces bt"
corn_bt_tidy.head()
```

```python
corn_herb_tidy = pd.melt(corn_herbicide,
                         id_vars=["State"],
                         var_name="Years",
                         value_name="Percent of Crop").dropna()
corn_herb_tidy["Crop"] = "Corn"
corn_herb_tidy["Variety"] = "Herbicide Resistant"

corn_herb_tidy.head()
```

```python
corn_all = pd.concat([corn_herb_tidy, corn_bt_tidy])
corn_all.head()
```

```python
corn_all["Crop Type"] = "Genetically Engineered (GE) Corn"
corn_all.head()
```

```python
corn_all["Percent of Crop"] = corn_all["Percent of Crop"].astype("int")
```

```python
State = "Iowa"
Variety = "Produces bt"

corn_all[(corn_all["State"] == State) & (corn_all["Variety"] == Variety)].plot(
                                                                    kind="line",
                                                                    x="Years",
                                                                    y="Percent of Crop")
```
