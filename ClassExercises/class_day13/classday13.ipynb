```python
import pandas as pd
import numpy as np
```

```python
PopGe = pd.read_csv("Ptrichocarpa_444_v3.1.annotation_info.txt", sep="\t")
PopGe.head()
```

```python
GO_Terms = pd.read_csv("GO.terms.txt", sep="\t", header=None, names=["Vocabulary", "Term", "Definition"])
GO_Terms.head()
```

```python
PopGe = (PopGe[["peptideName", "GO"]])
PopGe = PopGe.dropna(axis=0)
PopGe.head()
```

```python
PopGe["GO"].str.split(pat=", ")
NewPopGe = []

for index, row in PopGe.iterrows():
    Go_terms = row.GO.split(", ")
    Name_Len = np.repeat(row.peptideName, len(Go_terms))
    df =pd.DataFrame({"peptideName": Name_Len, "GO" : Go_terms})
    NewPopGe.append(df)
NewPopGe.head()
```

```python
poplar_go = pd.concat(NewPopGe)
poplar_go.head()
```
