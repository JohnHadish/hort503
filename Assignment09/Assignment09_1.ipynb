```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

%matplotlib inline

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
```

```python
plt.plot(X,C)
plt.plot(X,S)

plt.show()
```

```python
#Make a new figure of size 8 by 6 with pixel ratio 100 per square inch? Why is 100?
plt.figure(figsize=(8,6), dpi=80)

#Makes a new subplot 1 x 1
plt.subplot(111)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

# Plot cosine using blue color with a continuous line width 1 pixel
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")


#Ploy sin using green with a continouso line widthe 1 pixel
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")

#Set x limits
plt.xlim(-4.0,4.0)

#Set x ticks
plt.xticks(np.linspace(-4,4,9,endpoint=True))

#Set y ticks
plt.yticks(np.linspace(-1,1,5,endpoint=True))

#Save figure using 72 dots per inch
plt.savefig("exercise2.png",dpi=72)

#Show results on screen
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi=80)

#makes the lines have color, width, linestyle and a label
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sine")

#this plots a legend in the upper left box. It plots the above label 
plt.legend(loc="upper left", frameon=False)

#this makes our label tick marks look very nice. The first part of each one
#is where they are to be put, and the second part uses latex to make them 
#look nice
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,+1],
          [r'$-1$', r'$0$', r'$+1$'])

#Now we are going to do some annotations:
t = 2*np.pi/3

#How to plot lines is kind of goofy. The first tuple is x axis for start and end, and the second is y axis for
#start and end
plt.plot([t,t], [0, np.cos(t)], color='blue', linewidth=1.5, linestyle="--")
plt.scatter([t,],[np.cos(t),], 50, color='blue')

plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
            xy=(t, np.sin(t)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3, rad=.2"))

plt.plot([t, t], [0, np.sin(t)], color="red", linewidth=1.5, linestyle="--")
plt.scatter([t,],[np.cos(t),], 50, color="red", linewidth= 1.5, linestyle="--")

plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
            xy=(t, np.cos(t)), xycoords='data',
            xytext=(-90,-50), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

#This plots spines (the lines in the middle vs the side
#This removes the right and top
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#This moves the bottom and left spines to the middle
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))

plt.show()
```

```python
#Scatter plot
n = 1024

# np.normal does Gausian centered around first number
X = np.random.normal(0,5,n)
Y = np.random.normal(0,5,n)
T = np.arctan2(Y,X)
print(X)
print(T)

plt.xticks([]), plt.yticks([])


plt.scatter(X, Y, s=75, c=T, alpha =.5 )
plt.show()
```

```python
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n=1000
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.figure(figsize=(20,15))
plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='hot')
C = plt.contour(X, Y, f(X,Y), 8, colors='black', linewidths=2)

plt.clabel(C, inline=1, fontsize=12)

plt.xticks([]), plt.yticks([])


plt.show()
```
